```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
%matplotlib inline
```

```python
# line graph
x = np.arange(-3,3,0.001)
plt.plot(x,norm.pdf(x)) # probability density function
plt.show()
```

```python
# multiple line gaph

x = np.arange(-3,3,0.001)
# y = np.arange(-4,4,0.002)
plt.xlabel('Greebles')
plt.ylabel('Probabilities')
plt.plot(x,norm.pdf(x),'b-') # blue solid line
plt.plot(x,norm.pdf(x,1,0.5),'r--') # style of line
plt.plot(x,norm.pdf(x,-2,0.5),'g:')

plt.legend(['Sneetches','Gacks','Gusoams'],loc=0)
plt.show()
```

```python
#Pie Chart

values = [12,55,4,14]
colors = ['r','g','b','c','m']
explode = [0,0.1,0,0]
labels = ['Lahore','Faislabad','Karachi','Islamabad']
plt.pie(values,colors=colors,labels=labels,explode=explode)
plt.title("Illitracy")
plt.show()
```

```python
# # bar chart
# plt.bar(x-axis,y-axis,colors)
values = [12,55,4,14]
colors = ['r','g','b','c','m']
plt.bar(range(0,4),values,color = colors)
plt.show()
```

```python
# scatter plot
X = np.random.randn(500) #normal distribution
Y = np.random.rand(500)
plt.scatter(X,Y)
plt.show()
```

```python
# histogram
incomes = np.random.normal(3000,15000,5000)
# plt.hist(x-axis)
plt.hist(incomes,bins=40)
plt.show()
```

```python
#box plot
x = np.random.randint(0,80,100)
plt.boxplot(x)
plt.show()
```
